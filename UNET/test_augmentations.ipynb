```python
import numpy as np
import os
from matplotlib import pyplot as plt
from PIL import Image
```

```python
train_dir = '../dataset/BBBC_039_formatted/train/'

# get a random image from train_dir
img_path = np.random.choice(os.listdir(train_dir + 'images/' ))
img = Image.open(train_dir + 'images/' + img_path)
mask_path = img_path.replace('.tif', '.png')
mask = Image.open(train_dir + 'masks/' + mask_path)

img = np.array(img)
mask = np.array(mask)
```

```python
plt.imshow(img)
```
